```python
import pandas as pd
import numpy as np

df = pd.read_csv("mock_payer_data.csv")

null_counts = df.isnull().sum()
duplicate_count = df.duplicated().sum()
high_outliers = df[df['claim_amount'] > 2000000]
high_outlier_count = len(high_outliers)
df["claim_date_converted"] = pd.to_datetime(df["claim_date"], errors="coerce")
wrong_claim_date_count = df["claim_date_converted"].isna().sum()
df["dob_converted"] = pd.to_datetime(df["dob"], errors="coerce")
wrong_dob_count = df["dob_converted"].isna().sum()

print("\n==================== DATA QUALITY SUMMARY ====================\n")
print("1. Null Count For Each Column:")
print(null_counts)
print("\n---------------------------------------------------------------\n")
print(f"2. Total Duplicate Rows: {duplicate_count}")
print("\n---------------------------------------------------------------\n")
print(f"3. Claim Amount > 20 LAK Count: {high_outlier_count}")
print("\n---------------------------------------------------------------\n")
print(f"4. Wrong Format Claim Dates: {wrong_claim_date_count}")
print("\n---------------------------------------------------------------\n")
print(f"5. Wrong Format DOB Values: {wrong_dob_count}")
print("\n===============================================================\n")
```
