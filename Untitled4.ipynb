```python
df = df.drop_duplicates()
```

```python
df['claim_amount'] = df['claim_amount'].fillna(df['claim_amount'].median())
df['member_name'] = df['member_name'].fillna("Unknown")
```

```python
df = df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df['claim_date'] = pd.to_datetime(df['claim_date'], errors='coerce')
```

```python
df[df['claim_date'].isna()]
```

```python
high_outliers = df[df['claim_amount'] > 2000000]
high_outliers
print(len(high_outliers))

df['claim_date_converted'] = pd.to_datetime(df['claim_date'], errors='coerce')
wrong_format_count = df['claim_date_converted'].isna().sum()
print(wrong_format_count)

df['claim_date_converted'] = pd.to_datetime(df['dob'], errors='coerce')
wrong_format_count = df['claim_date_converted'].isna().sum()
print(wrong_format_count)
```

```python
from sklearn.ensemble import IsolationForest

model = IsolationForest(contamination=0.03, random_state=42)
df['Anomaly'] = model.fit_predict(df[['claim_amount']])

df_anomalies = df[df['Anomaly'] == -1]
df_anomalies.head()
```
